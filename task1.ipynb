1) Un desarrollador junior de su equipo sugiere: "Para detectar con mayor precisión las texturas de las
hojas enfermas, deberíamos construir una red secuencial clásica (tipo VGG) pero de 150 capas. Más
profundo siempre es mejor". Como líder técnico, explíquele argumentativamente por qué esta red
fracasará estrepitosamente en el entrenamiento (mencionando el fenómeno de degradación y el
desvanecimiento del gradiente). Luego, justifique cómo la adición estructural de las conexiones
residuales (𝐹(𝑥) + 𝑥) de ResNet rescata el proyecto, haciendo viable entrenar redes ultra-profundas
sin colapsar.

Me gusta tu sugerencia entiendo tu logica, pero en la practica meterle 150 capas a una red tipo VGG es cavar nuestra propia tumba por dos razones principales, una es el desvanecimiento de la gradiente ya que para que la red aprenda usamos el gradiente que viaja desde la ultima capa hasta la primera y aqui el problema es que en una red secuencial tan larga ese valor se va multiplicando por numeros muy pequeños en cada capa y cuando llega a las primeras capas que son las que deberian aprender las texturas basicas de las hojas el gradiente es practicamente cero. La red deja de aprender y el entrenamiento se queda estancado.

La otra razón es por un fenomeno de degradación que uno pensaría que si la red no mejora al menos se queda igual pero no, en estos casos donde se manejan redes tan profundas y simples el error de entrenamiento empieza a subir. No es que haya overfitting sino es que el sistema se vuelve tan complejo matematicamente que el optimizador ni siquiera puede encontrar una solución decente y basicamente la red colapsa bajo su propio peso.

Aqui para rescatar el proyecto no necesitamos menos capas sino conexiones residuales que es donde entra la La magia de ResNet que es que no obliga a cada capa a aprender todo desde cero, en lugar de que una capa intente adivinar la imagen completa solo tiene que aprender la diferencia entre lo que ya sabe y lo que falta por mejorar.

Esto nos salva ya que hablando de Gradientes de fluidos el error puede viajar por esos atajos sin perderse, llegando a las capas iniciales con fuerza y se genera una identidad garantizada donde si una capa no tiene nada nuevo que aportar simplemente deja pasar la información lo que hace que entrenar una red de muchas capas sea viable, permitiendonos detectar esos detalles finos de las enfermedades sin que el modelo explote.

2. Las enfermedades en las hojas de mango son visualmente heterogéneas: La Antracnosis se
presenta como puntos negros diminutos, mientras que el Moho Polvoriento cubre áreas enormes de
la hoja. Analice cómo la topología en paralelo del módulo Inception (usando filtros 3x3 y 5x5
simultáneamente) es ideal para este problema biológico en particular. Además, desde una
perspectiva de costos de infraestructura (uso de GPUs en AWS o Google Cloud), explique cómo la
inserción estratégica de convoluciones de 1x1 evita la explosión de la dimensionalidad y salva el
presupuesto mensual de la startup.

Para el caso de la Antracnosis como la enfermedad son putos negros dinimunots si necesitamos una lupa que en este caso se adecua mas el filtro de 3x3 que se enfoque en detalles muy localizados y por otra parte el filtro de 5x5 seria mejor para el Moho Polvoriento ya que es una mancha mas grande, si solo usariamos filtros pequeños la red no veria el panorama completo lo que aqui nos ayuda este filtro de 5x5 que actua como una captura panoramica que podra ver los patrones de esta enfermedad.

Al usar el modulo inception no se tendra que elgir uno o el otro ya que la red procesa la imagen con los 2 filtros al mismo tiempo en paralelo y junta los resultados para que el modelo sea capaz de reconocer tanto los puntos pequeños como las manchas.

Ahora para el tema del presupuesto si conectamos las capas de 3x3 y de 5x5 directamente el numero de parametros y el costo de computo es muy alto por lo que aqui podriamos meter un cuello de botella que se basa en pasar la imagen en una convolucion de 1x1 antes de pasarla por los filtros, la primera capa funcionaria como un compresor de archivos que reduciria la cantidad de canales de informacion sin perder caracteristicas necesarias lo que se traduciria en procesar menos datos que harian que las gpus trabajen menos y por lo tanto el costo bajaria.

3. El cliente final (el agricultor) usará la aplicación en un teléfono Android de gama baja en medio del campo, sin conexión a internet. Sabemos que MobileNet logra esta eficiencia gracias a la Depthwise
Separable Convolution. Describa brevemente cómo esta convolución divide el trabajo (filtrado
espacial vs. combinación de canales). Sin embargo, en ingeniería no hay soluciones mágicas, todo tiene un precio: ¿Qué capacidad matemática o expresividad de la red se está sacrificando al separar
estos dos pasos en comparación con una convolución estándar? Analice críticamente por qué, como
director del proyecto, usted acepta pagar ese "costo" en este escenario comercial.

En este caso en una convolucion se intenta hacer todo a la vez pero MobileNet es inteligente y divide el trabajo en dos partes:

- Depthwise Convolution: que en lugar de usar un filtro que mire todos los canales usa un filtro individual para cada canal y solo se encarga de entender las formas en ese canal especifico.

- Pointwise Convolution: Luego usa filtros de 1x1 para unir la informacion de todos esos canales y crear nuevas caracteristicas.

Pero al separar los pasos estamos limitando la capacidad de representacion de la red que en una convolucion estandar la red puede aprender relaciones muy complejas entre el espacio y el color al mismo tiempo pero al separarlas la red se vuelve mas sencilla matematicamente lo que se traduce en que MobileNet podra tener un poco mas de problemas para distinguir detalles sutiles que una red robusta si podria ver.

Y como Director de proyecto el costo lo aceptaria ya que basicamente si un modelo aunque sea perfecto no puede correr no va a servir para nada, considerando el contexto del agricultor esta seria la forma de llevar a cabo el proyecto.
